```python
import numpy as np
import matplotlib.pyplot as plt
from wavelet import compute_cwt
```

### Sine function without noise

```python
dt = 1 / 1e3 ## interval between two consecutive time
t = np.arange(1000)*dt
f1 = 40
x = 5*np.sin(2 * np.pi * f1 * t) 
```

```python
plt.plot(t, x)
plt.xlabel("t");
```

```python
cwtm, cwtm_sig, freq = compute_cwt(x, t)
```

```python
plt.pcolormesh(t, freq, cwtm, cmap='Spectral_r')
plt.colorbar()
plt.xlabel('Time (s)')
plt.ylabel('Freq (Hz)')
```

### Two sine functions without noise

```python
dt = 1 / 1e3 ## interval between two consecutive time
t = np.arange(1000)*dt
f1 = 40
f2 = 100
x = 3*np.sin(2 * np.pi * f1 * t[:500]) 
x = np.append(x, 2*np.sin(2 * np.pi * f2 * t[:500]))
```

```python
cwtm, cwtm_sig, freq = compute_cwt(x, t)
```

```python
plt.pcolormesh(t, freq, cwtm, cmap='Spectral_r')
plt.colorbar()
plt.xlabel('Time (s)')
plt.ylabel('Freq (Hz)')
```

### Two sine functions with noise

Significance against AR1 Process of time series is given where cwtm_sig >= 1.

```python
dt = 1 / 1e3 ## interval between two consecutive time
t = np.arange(1000)*dt
f1 = 40
f2 = 100
x = 3*np.sin(2 * np.pi * f1 * t[:500]) 
x = np.append(x, 2*np.sin(2 * np.pi * f2 * t[:500]))
x = x + np.random.normal(0,1,1000)
```

```python
plt.plot(t, x)
plt.xlabel("t");
```

```python
cwtm, cwtm_sig, freq = compute_cwt(x, t)
```

```python
plt.pcolormesh(t, freq, cwtm, cmap='Spectral_r')
plt.colorbar()
# add significance as countour
plt.contour(t, freq, cwtm_sig, [1], colors='k', zorder=1)
plt.xlabel('Time (s)')
plt.ylabel('Freq (Hz)')
```
