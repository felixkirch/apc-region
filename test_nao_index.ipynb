```python
import numpy as np
import matplotlib.pyplot as plt
from wavelet import compute_cwt
```

```python
# NOA Index Data from NOAA
# https://www.cpc.ncep.noaa.gov/data/teledoc/telecontents.shtml

data = np.loadtxt("data/tele_index.nh", skiprows = 19, usecols=[0,1,2])
```

```python
x =  data[:, 2]
t = np.arange(len(x)) # in month since 1950
```

```python
plt.plot(t, x)
plt.title("NAO Index")
plt.xlabel("t");
```

```python
cwtm, cwtm_sig, freq = compute_cwt(x, t)
```

```python
plt.pcolormesh(t/12, freq[1:], cwtm_sig[1:, :], cmap='Spectral_r')
plt.colorbar()
# add significance as countour
plt.contour(t/12, freq[1:], cwtm_sig[1:, :], [1], colors='k', zorder=1)
plt.xlabel('t (years)')
plt.ylabel('f (1/month)');
```
